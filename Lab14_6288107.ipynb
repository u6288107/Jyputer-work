# Exercise 11

```python
import numpy as np
test_image = img. copy ()
lx , ly , lz = test_image . shape
X, Y = np. ogrid [0:lx , 0: ly]
mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 4;
test_image [ mask*2 ]=0;
io. imshow ( test_image )
io. show ()
```

# Exercise 16

```python
from skimage import filters
blurred_img = filters . gaussian (img , sigma =3 , multichannel =
True )
io. imshow ( blurred_img )
io. show ()
```
